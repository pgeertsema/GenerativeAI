# Hands on : optimal compute #

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatLogSlider
import ipywidgets as widgets

# Define the scaling exponents based on the paper's values
alpha_S = 0.076
alpha_T = 0.095
alpha_C = 0.05

# Baseline values for model size, tokens, and compute
S_base = 1e9  # example baseline model size
T_base = 1e12  # example baseline number of tokens
C_values = np.logspace(9, 30, 100)  # extended compute range on a log scale

# Define a large, fixed range for the y-axis
y_min, y_max = 0, 0.01  # Example fixed range, adjust as needed

# Calculate test loss as a function of model size (S), tokens (T), and compute (C)
def compute_loss(S, T):
    L = (S ** -alpha_S) * (T ** -alpha_T) * (C_values ** -alpha_C)
    return L

# Create the interactive plot
def plot_loss(S, T):
    # Compute the test loss for given S and T
    L_values = compute_loss(S, T)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(C_values, L_values, label=f"Model Size: {S:.1e}, Tokens: {T:.1e}")
    plt.axvline(x=1e26, color='red', linestyle='--', label="Compute = $10^{26}$")
    plt.xscale('log')
    plt.ylim(y_min, y_max)  # Fix the y-axis range
    plt.xlabel("Total Compute (C)")
    plt.ylabel("Test Loss (L)")
    plt.title("Test Loss vs. Total Compute (Fixed Y-axis Range)")
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()

# Define interactive sliders for model size (S) and tokens (T)
S_slider = FloatLogSlider(value=S_base, base=10, min=8, max=11, step=0.1, description="Model Size (S)")
T_slider = FloatLogSlider(value=T_base, base=10, min=10, max=14, step=0.1, description="Tokens (T)")

# Set up the interactive plot using ipywidgets
interact(plot_loss, S=S_slider, T=T_slider)
```
